### Условие на заданието
<br>
Даден е csv файл (email_spam_classification.csv) с извадка данни, представяща в табличен вид
броя срещания на предварително указани думи на английски език (по колоните) в конкретни
имейл съобщения - всеки ред съотвества на отделен и независим от другите обработен текст
от имейл. Използвайки колоните от 1 до n-1, тренирайте класификационни алгоритми, с които
да предвидите стойността в последната колона n (explained variable), която съдържа
единствено стойности {1, 0}, указващи респективно дали съобщението е спам или не. 

### Решение:
<br>
За да могат да се класифицират мейлите на нормални и на спам с използвани два класификационни алгоритми, различни от невронна мрежа: 
<br>
1. k най-близките съседи (k-Nearest Neighbours)
<br>
2. Support Vector Machine(SVM)
<br>
Добавена е и невронна мрежа като допълнение към останалите алгоритми

### Етапи за решаване на проблема
<br>
1. Четене на данни и тяхната предварителна обработка:
<br>
Даден е csv файл с редове текстове на даден мейл, а за колони са думи от английския език, като е дадено колко пъти дадена дума се среща в даден мейл текст. В последната колона се съдържат стойности от 0 и 1 за да се покаже дали даденото съобщение е спам или нормален мейл. Този тип данни се нарича "bag of words".
<br>
Данните се прочитат използвайки библиотеката "pandas" и след това се отделят на "думи"(това са всички думи от файла) и резултатната колона "Спам"(в нея са данните дали даденото съобщение е спам или не).
<br>
След като се отделят думите, трябва да им се направи preprocesing за да се нормализират със стойности от 0 до 1(min = 0, max = 1).
<br>
<br>
2. Използване на алгоритми за класификация и k-Fold cross validation(където k > 5)
<br>
Целта за използването на повече от един алгоритъм за класификация е да се сравнят алгоритмите и да се види кой от тях е по-точен.
<br>
<br>
За метрика е използвана F1, която:
<br>
F1 = 2 * (precision * recall) / (precision + recall), където
<br>
precision = процента от спама, който сме открили
<br>
recall = процента, който наистина е спам от откритите данни за спам

### Използване на k-Nearest Neighbours

```python
#Solution with KNeighborsClassifier

#Import libraries
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score

#Load csv file with bag of words
email_spam_csv = pd.read_csv("C:\\Users\\garatlakov\\MSMO-Project\\data\\email_spam_classification.csv")
dataset = email_spam_csv.values
dataset = dataset[:, 1:]

words_columns = dataset[:, :-1]
prediction_column = dataset[:, -1]

words_columns = words_columns.astype('int');
prediction_column = prediction_column.astype('int')

#normalisation from 0 to 1
min_max_scaler = preprocessing.MinMaxScaler()
words_normalized = min_max_scaler.fit_transform(words_columns)

#Using KNeighborsClassifier
knn_estimator = KNeighborsClassifier(n_neighbors=3) 

#Evaluate a score by cross-validation where cv > 5
knn_cv_scores = cross_val_score(knn_estimator, words_normalized, prediction_column, cv = 6, scoring = "f1")

print(knn_cv_scores)
print("knn_cv_scores mean:{}".format(np.mean(knn_cv_scores)))
```

### Използване на Support Vector Machine Classifier

```python
#Solution with Support Vector Machine Classifier

#Import libraries
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn import svm
from sklearn.model_selection import cross_val_score

#Load csv file
email_spam_csv = pd.read_csv("C:\\Users\\garatlakov\\MSMO-Project\\data\\email_spam_classification.csv")
dataset = email_spam_csv.values
dataset = dataset[:, 1:]

words_columns = dataset[:, :-1]
prediction_column = dataset[:, -1]

words_columns = words_columns.astype('int');
prediction_column = prediction_column.astype('int')

#normalisation from 0 to 1
min_max_scaler = preprocessing.MinMaxScaler()
words_normalized = min_max_scaler.fit_transform(words_columns)

#Using Support Vector Machine Classifier
clf = svm.SVC()

#Evaluate a score by cross-validation where cv > 5
spv_cv_scores = cross_val_score(clf, words_normalized, prediction_column, cv = 6, scoring = "f1")

print(spv_cv_scores)
print("spv_cv_scores mean:{}".format(np.mean(spv_cv_scores)))
```

### Визуализиране на резултатите

```python
import matplotlib.pyplot as plt

plt.plot(knn_cv_scores)
plt.plot(spv_cv_scores)
plt.legend(['F1 score KNN', 'F1 score SVM'], loc="upper right")
plt.show()
```
